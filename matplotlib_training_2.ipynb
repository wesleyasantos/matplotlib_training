# Passeios aleatórios

### Criando a classe RandomWalk()

```python
from random import choice
```

```python
class RandomWalk():
    """Uma classe para gerar passeios aleatórios"""
    
    def __init__(self, num_points=5000):
        """Inicializa os atributos de um passeio"""
        self.num_points = num_points
        
        # Todos os passeios começam em (0, 0)
        self.x_values = [0]
        self.y_values = [0]
        
    # Escolhendo as direções
    def fill_walk(self):
        """Calcula todos os pontos do passeio"""
        
        # Continua dando passos até que o passeio alcance o tamanho desejado
        while len(self.x_values) < self.num_points:
            
            # Decide direção a ser seguida e distância a ser percorrida nessa direção
            # Resultados para x_step: Positivo = Direita | Negativo = Esquerda | 0 = Movimento Vertical
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
            
            # Resultados para y_step: Positivo = Cima | Negativo = Baixo | 0 = Movimento Horizontal
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            # Rejeita movimentos que não vão a lugar nenhum
            if x_step == 0 and y_step == 0:
                continue
                
            # Calcula os próximos valores de x e y
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step
            
            self.x_values.append(next_x)
            self.y_values.append(next_y)
```

### Plotando o passeio aleatório

```python
import matplotlib.pyplot as plt
```

```python
# Plota os pontos do RandomWalk()
rw = RandomWalk()
rw.fill_walk()

plt.scatter(rw.x_values, rw.y_values, s=1)
plt.show()
```

### Estilizando o passeio

```python
rw = RandomWalk()
rw.fill_walk()

# Usamos range para gerar uma lista de números igual ao número de pontos do passeio e armazenamos esses números em 'point_numbers'
# que usaremos para definir a cor de cada ponto do passeio e em seguida, passamos 'point_numbers' para 'c'.
# Usaremos o 'colormap Blues' e passamos 'edgecolor' como 'none' para eliminar o contorno preto de cada ponto.
point_numbers = list(range(rw.num_points))
plt.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolor='none', s=4)

# O resultado é um gráfico do passeio que varia do azul claro para o azul escuro, ao longo de um gradiente.
plt.show()
```

### Plotando os pontos de ínicio e fim

```python
rw = RandomWalk()
rw.fill_walk()

point_numbers = list(range(rw.num_points))
plt.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolor='none', s=4)

# Enfatiza o primeiro e o último ponto
plt.scatter(0, 0, c='green', edgecolor='none', s=20)
plt.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolor='none', s=20)

plt.show()
```

### Adicionando pontos para plotagem

```python
# Para aumentar o número de pontos para termos mais dados com os quais possamos trabalhar,
# basta aumentar o valor de 'num_points' na criação de uma instância de 'RandomWalk()'
rw = RandomWalk(50000)
rw.fill_walk()

point_numbers = list(range(rw.num_points))
plt.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolor='none', s=1)

plt.scatter(0, 0, c='green', edgecolor='none', s=20)
plt.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolor='none', s=20)

plt.show()
```

```python
# Uma visualização será eficaz para informar padrões em dados se couber apropriadamente na tela.
# Para deixar a janela de plotagem mais adequada à sua tela, ajuste o tamanho de saída do matplotlib.

rw = RandomWalk(50000)
rw.fill_walk()

# Define o tamanho da janela de plotagem
plt.figure(figsize=[10, 6])

point_numbers = list(range(rw.num_points))
plt.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolor='none', s=1)

plt.scatter(0, 0, c='green', edgecolor='none', s=30)
plt.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolor='none', s=30)

plt.show()
```
