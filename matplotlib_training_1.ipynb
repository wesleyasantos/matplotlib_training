```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
```

```python
plt.plot(input_values, squares, linewidth=2)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=14)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of value", fontsize=14)

# Define o tamanho dos rótulos das marcações
plt.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
# Scatter() -> Passe um úinico par (x, y) do ponto em que estiver interessado para scatter()
plt.scatter(2, 4)
plt.show()
```

```python
# Estilizando a saída - 's=200' -> Define o tamanho dos pontos
plt.scatter(2, 4, s=200)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=14)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Define o tamanho dos rótulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
# Plotando uma série de pontos com scatter
# AQUI, vamos passar listas separadas de valores de x e y para scatter()
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.scatter(x_values, y_values, s=50)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=14)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Define o tamanho dos rótulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)

# Um gráfico de dispersão com vários pontos
plt.show()
```

```python
# Calculando dados automaticamente
x_values1 = list(range(1, 1001))

# -> List Comprehension: Para cada valor de x em x_values1, retorne x**2
y_values1 = [x**2 for x in x_values1] 
```

```python
# -> edgecolor: remove o contorno dos pontos
plt.scatter(x_values1, y_values1, s=5, edgecolor='none') 


# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=14)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Define o tamanho dos rótulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)

# Define o intervalo para cada eixo
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
# Para mudar a cor dos pontos, passe 'c' para scatter() com o nome de uma cor a ser usada:

# -> edgecolor: remove o contorno dos pontos
plt.scatter(x_values1, y_values1, s=5, edgecolor='none', c='red') 


# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=14)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Define o tamanho dos rótulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)

# Define o intervalo para cada eixo
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
# Usando 'colormap': Usado em visualizações para enfatizar um padrão nos dados

plt.scatter(x_values1, y_values1, s=5, edgecolor='none', c=y_values1, cmap=plt.cm.Blues) 


# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=14)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Define o tamanho dos rótulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)

# Define o intervalo para cada eixo
plt.axis([0, 1100, 0, 1100000])

plt.show()
```
